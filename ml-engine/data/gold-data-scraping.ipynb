```python
import yfinance as yf
import pandas as pd
from datetime import datetime

ticker = "GC=F"
name = "Gold Futures"
symbol = "GOLD"
today = datetime.today()

ohlc = yf.download(ticker, period="2y", interval="1d")[['Open', 'High', 'Low', 'Close']]
ohlc.dropna(inplace=True)
ohlc.to_csv("./historical-data/gold.csv")
print("-> Saved OHLC data to gold.csv")

# --- Historical Close for % Change Reference ---
hist = yf.Ticker(ticker).history(period="max")['Close']
hist.dropna(inplace=True)

# --- Helper: Calculate % change from most recent close ---
def get_change(ref_days):
    ref_date = hist.index[-1] - pd.Timedelta(days=ref_days)
    ref_value = hist[:ref_date].iloc[-1] if not hist[:ref_date].empty else None
    latest = hist.iloc[-1]
    return round(((latest - ref_value) / ref_value) * 100, 2) if ref_value else None

# --- Calculate % Changes ---
change_df = pd.DataFrame([{
    "Symbol": symbol,
    "1W Change %": get_change(7),
    "1M Change %": get_change(30),
    "3M Change %": get_change(90),
    "6M Change %": get_change(180),
    "1Y Change %": get_change(365),
    "5Y Change %": get_change(365*5),
    "Name": name
}])

change_df.to_csv("./market-data/gold-market-data.csv", index=False)
print("-> Saved '%' change data to gold-market-data.csv")
```
